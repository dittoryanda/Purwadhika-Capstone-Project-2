## Data Cleansing

### Menghapus kolom yang semua value-nya kosong.

```python
df.drop(['Unnamed: 21','Unnamed: 22','Unnamed: 23','Unnamed: 24','Unnamed: 25'], axis = 1, inplace = True)
```

```python
df.info()
```

### Menghapus semua baris yang value-nya kosong

```python
df.dropna(axis=0, thresh=1, inplace=True)
```

### Kolom 'grand_total' dan ' MV '

```python
df.sample(25)[['price','qty_ordered','discount_amount','grand_total',' MV ','sales_commission_code']]
```

Dari hasil pengamatan sampel, column 'grand_total' tidak konsisten dengan perhitungan qty x price - discount. Terdapat kolom yang lebih sesuai dengan perhitungan, yaitu kolom 'MV', namun kolom tersebut bertipe object dan tidak menyimpan bilangan koma desimal. Oleh karena itu, kolom baru perlu dibuat yang berisi hasil perhitungan rumus qty x price - discount.

```python
df.drop(['grand_total', ' MV '], axis=1, inplace=True)
df
```

```python
df['total'] = df['price']*df['qty_ordered']-df['discount_amount']
```

Terdapat beberapa baris yang totalnya dibawah 0. Kita akan drop baris-baris tersebut karena tidak valid.

```python
df[df['total']<0]
```

```python
df.drop(df[df['total']<0].index, inplace=True)
```

### Kolom 'Working Date' dan 'created_at'

```python
df['Working Date'].equals(df['created_at'])
```

Karena kolom 'Working Date' dan 'created_at' sama, maka salah satu kolom bisa di-drop

```python
df.drop(['Working Date'], axis = 1, inplace=True)
```

Kolom 'created_at' bertipe object maka kita harus mengganti tipe datanya ke datetime.

```python
df['created_at']=pd.to_datetime(df['created_at'])
```

### Kolom 'Month', 'Year', 'M-Y' dan berisi bulan dan tahun dari kolom 'created_at' sehingga bisa di-drop

```python
df.drop(['Month','Year','M-Y'], axis=1, inplace=True)
```

### Menampilkan informasi unique value setiap column-nya

```python
pd.set_option('display.max_colwidth', None)
listItem = []
for col in df.columns :
    listItem.append( [col, df[col].nunique(), df[col].unique()])

tabel1Desc = pd.DataFrame(columns=['Column', 'Unique', 'Values'],
                     data=listItem)
tabel1Desc
```

### Kolom 'increment_id', 'item_id', 'sales_commission_code', 'FY', dan 'sku'

Kolom-kolom ini kurang memberi informasi yang penting sehingga bisa di-drop. Dalam hal ini yang dimaksud, isi value dari column-column di atas biasanya merupakan foreign key mengacu ke sebuah tabel di database (dalam kasus ini tidak ada sehingga dapat diabaikan).

Kita menggunakan kolom 'increment_id' sebagai parameter, karena setiap order hanya memiliki satu increment id yang unique.

### Handle missing value pada kolom 'status' dan 'category_name_1'

```python
df[df['status'].isna()]
```

Pada kolom 'status', value kosong akan diisi dengan value yang paling sering muncul di kolom 'status' dengan kondisi metode pembayaran dengan cod dan isi value BI status gross karena setelah dilihat baris dengan value kolom 'status'-nya kosong, semua metode pembayarannya menggunakan cod dan BI status berisi gross.

```python
df['status'] = df['status'].fillna(df[(df['payment_method']=='cod') & (df['BI Status']=='Gross')]['status'].mode()[0])
```

```python
df.dropna(subset='category_name_1', axis=0, inplace=True)
```

## Data Analysis

### Harga total setelah diskon yang di-order pada e-commerce dikategorikan berdasarkan tahun.

```python
x1 = df['created_at'].dt.year.unique()
x1_list = x1.tolist()
x1_list
```

```python
y1 = df['total'].groupby(df['created_at'].dt.year).sum()
y1_list = y1.tolist()
y1_list
```

```python
df_bar = pd.DataFrame(list(zip(x1_list,y1_list)), columns=['year', 'total'])
df_bar
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data = df_bar, x='year', y='total')
```

### Melihat total penjualan setiap tahunnya berdasarkan kategori

```python
values2016 = df[df['created_at'].dt.year==2016.0].groupby(['category_name_1'])['total'].sum().sort_values(ascending=False)
x = values2016.keys().tolist()
y2016 = values2016.tolist()
df_2016 = pd.DataFrame(list(zip(x,y2016)), columns=['category_name_1', 'total'])
plt.figure(figsize=(12,4))
plt.xticks(rotation=45)
sns.barplot(data = df_2016, x='category_name_1', y='total', palette='muted')
plt.title('2016 Total By Categories')
plt.show()

values2017 = df[df['created_at'].dt.year==2017.0].groupby(['category_name_1'])['total'].sum().sort_values(ascending=False)
x = values2017.keys().tolist()
y2017 = values2017.tolist()
df_2017 = pd.DataFrame(list(zip(x,y2017)), columns=['category_name_1', 'total'])
plt.figure(figsize=(12,4))
plt.xticks(rotation=45)
sns.barplot(data = df_2017, x='category_name_1', y='total', palette='muted')
plt.title('2017 Total By Categories')
plt.show()

values2018 = df[df['created_at'].dt.year==2018.0].groupby(['category_name_1'])['total'].sum().sort_values(ascending=False)
x = values2018.keys().tolist()
y2018 = values2018.tolist()
df_2018 = pd.DataFrame(list(zip(x,y2018)), columns=['category_name_1', 'total'])
plt.figure(figsize=(12,4))
plt.xticks(rotation=45)
sns.barplot(data = df_2018, x='category_name_1', y='total', palette='muted')
plt.title('2018 Total By Categories')
plt.show()
```

### Melihat total penjualan berdasarkan status order

```python
status2016 = df[df['created_at'].dt.year==2016.0].groupby(['status'])[['total']].sum().sort_values(by='total', ascending=False)
plt.figure(figsize=(12,4))
plt.xticks(rotation=45)
sns.barplot(data = status2016, x=status2016.index, y='total', palette='muted')
plt.title('2016 Total By Status')
plt.show()

status2017 = df[df['created_at'].dt.year==2017.0].groupby(['status'])[['total']].sum().sort_values(by='total', ascending=False)
plt.figure(figsize=(12,4))
plt.xticks(rotation=45)
sns.barplot(data = status2017, x=status2017.index, y='total', palette='muted')
plt.title('2017 Total By Status')
plt.show()

status2018 = df[df['created_at'].dt.year==2018.0].groupby(['status'])[['total']].sum().sort_values(by='total', ascending=False)
plt.figure(figsize=(12,4))
plt.xticks(rotation=45)
sns.barplot(data = status2018, x=status2018.index, y='total', palette='muted')
plt.title('2018 Total By Status')
plt.show()
```

### Melihat total berdasarkan hari

```python
values = df.groupby(df['created_at'].dt.day_name()).sum()['total']
x = values.keys().tolist()
y = values.tolist()
df_line = pd.DataFrame(list(zip(x,y)), columns=['Day', 'Total'])
df_line
# sns.lineplot(data=df_line, x='Day', y='Total')
```

```python
df_line_arranged = pd.DataFrame(columns=['Day','Total'])

sunday = df_line[df_line['Day']=='Sunday']
monday = df_line[df_line['Day']=='Monday']
tuesday = df_line[df_line['Day']=='Tuesday']
wednesday = df_line[df_line['Day']=='Wednesday']
thursday = df_line[df_line['Day']=='Thursday']
friday = df_line[df_line['Day']=='Friday']
saturday = df_line[df_line['Day']=='Saturday']
df_line_arranged = df_line_arranged.append(sunday,ignore_index =True)
df_line_arranged = df_line_arranged.append(monday,ignore_index =True)
df_line_arranged = df_line_arranged.append(tuesday,ignore_index =True)
df_line_arranged = df_line_arranged.append(wednesday,ignore_index =True)
df_line_arranged = df_line_arranged.append(thursday,ignore_index =True)
df_line_arranged = df_line_arranged.append(friday,ignore_index =True)
df_line_arranged = df_line_arranged.append(saturday,ignore_index =True)


df_line_arranged
```

```python
x = df_line_arranged['Day']
y = df_line_arranged['Total']
plt.plot(x,y)
plt.title('Total by day')
```
